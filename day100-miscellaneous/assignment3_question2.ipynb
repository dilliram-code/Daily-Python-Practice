#### Assignment 3: Question 2

- Implement LDA using sklearn.discriminant_analysis

- Plot projected points on LDA axis

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis


# Create Dataset
# Feature matrix (X)
X = np.array([
    [2, 3],
    [3, 4],
    [4, 3],
    [3, 2],
    [7, 8],
    [8, 9],
    [9, 8],
    [8, 7]
])

# Labels
# 0 = Healthy
# 1 = Diseased
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Train LDA

lda = LinearDiscriminantAnalysis(n_components=1)

X_lda = lda.fit_transform(X, y)

# Display Results

print("Original Dataset:\n")
print(X)

print("\nLabels:\n")
print(y)

print("\nProjected Data (1D):\n")
print(X_lda)

print("\nClass Means:")
print(lda.means_)

print("\nLDA Coefficients (Projection Direction):")
print(lda.coef_)

print("\nExplained Variance Ratio:")
print(lda.explained_variance_ratio_)

# Plot Original Dataset

plt.figure(figsize=(6,6))

plt.scatter(
    X[y==0,0],
    X[y==0,1],
    color='blue',
    marker='o',
    s=100,
    label='Healthy'
)

plt.scatter(
    X[y==1,0],
    X[y==1,1],
    color='red',
    marker='^',
    s=100,
    label='Diseased'
)

plt.title("Original Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.legend()

plt.show()

# Plot LDA Projection

plt.figure(figsize=(10,2))

plt.scatter(
    X_lda[y==0],
    np.zeros(sum(y==0)),
    color='blue',
    marker='o',
    s=100,
    label='Healthy'
)

plt.scatter(
    X_lda[y==1],
    np.zeros(sum(y==1)),
    color='red',
    marker='^',
    s=100,
    label='Diseased'
)

plt.title("LDA Projection onto One Dimension")
plt.xlabel("LDA Component")
plt.yticks([])
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
