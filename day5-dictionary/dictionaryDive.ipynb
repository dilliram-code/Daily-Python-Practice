**❇️ What is a Dictionary in Python?**

A dictionary is a built-in hash map (key-value store) where:

- Keys must be unique and hashable

- Values can be any Python object

- Lookup, insert, delete = O(1) average time

```python
person = {
    "name": "Dilliram",
    "age": 30,
    "skills": ["Python", "ML", "DS"],
}
print(person)
```
