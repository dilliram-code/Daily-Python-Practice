**🔹 1. What is the requests module?**

- It’s a third-party library (not part of Python standard library).
- Used for sending HTTP/HTTPS requests like:

`GET (fetch data)`

`POST (send data)`

`PUT (update data)`

`DELETE (remove data)`

- Think of it as Python’s way of talking to websites and APIs.

```python
# Making Your First Request (GET)

import requests

response = requests.get("https://jsonplaceholder.typicode.com/posts/1")
print(response)
```

```python
import requests

response = requests.get("https://www.dilliramchaudhary.com.np/")
print(response)
```

```python
print(response.status_code)  # HTTP status (200 = OK)
print(response.text)  			# Raw text response
print(response.json)   			# sponse is JSON
```

**Explanation:**

`requests.get(url)` → sends a GET request.

`response` → object that holds all info about the request/response.

`status_code` → tells if it was successful (200 OK, 404 Not Found, etc.).

`text` → raw content of the response.

`json()` → converts response (if JSON) into a Python dict/list.

```python
# Sending Data with POST

import requests

url = "https://jsonplaceholder.typicode.com/posts"
data = {
	"title": "Hello World!",
  "body": "This is my first project.",
  "userId" : 1
  }

response = requests.post(url, data)

print(response.status_code)
print(response.json())
```

**Explanation:**

`requests.post(url, json=data)`→ sends data as JSON to the server.

- Many APIs expect JSON (key-value format).

- The server may return the created object back.
