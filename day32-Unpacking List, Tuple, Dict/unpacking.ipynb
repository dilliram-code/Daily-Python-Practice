#### Basic Unpacking

```python
nums = [1, 2, 3]
a, b, c = nums

print(a)  # 1
print(b)  # 2
print(c)  # 3
```

```python
# use * to store remaining values

nums = [10, 20, 30, 40, 50]
a, *rest = nums

print(a)     # 10
print(rest)  # [20, 30, 40, 50]
```

```python
# Middle unpacking

a, *mid, b = [1, 2, 3, 4, 5]

print(a)    # 1
print(mid)  # [2, 3, 4]
print(b)    # 5
```
