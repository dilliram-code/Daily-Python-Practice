````python
# basic get request
import requests

url = "https://api.github.com"
response = requests.get(url)

print(response.status_code)               # 200 means success
print(response.json())                    # Convert JSON response to Python dict
````

````python
# Fetch Data from Public API (Example: Random Joke API)

import requests

url = "https://official-joke-api.appspot.com/random_joke"
response = requests.get(url)

data = response.json()
print(data)

print(response.status_code)
print(f"{data['setup']} - {data['punchline']}")
print("----------------------------------")
print(f"{data['type']}\n{data['setup']}\n{data['punchline']}\n{data['id']}")
````

````python
# GET Request with Query Parameters
import requests
url = "https://api.agify.io"

response = requests.get(url)
print(response.status_code)
print(response.json())

print("----------------------------------------------------")

params = {"name": "dilli"}
response = requests.get(url, params=params)
print(response.json())

data = response.json()
print("------------------------------------------------------")
print(f"{data['count']}\n{data['age']}\n{data['name']}")
````

````python
# POST Request (Sending Data)

import requests
url = "https://httpbin.org/post"

payload = {"username": "dilli", "password": "1234"}
response = requests.post(url, json=payload)
print(response.json())
````
