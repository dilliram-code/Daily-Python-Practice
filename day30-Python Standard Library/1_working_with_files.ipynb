```python
# import pathlib lib
from pathlib import Path
```

```python
# create a path 
p = Path("data/file.txt")
print(p)

# this does not create the file, it just represents the path.
```

```python
# absolute vs relative path

Path("data/file.txt")                 # relative path
Path("/home/user/file.txt")           # absolute path        
```

```python
# get absolute path
p.resolve()
```

```python
# get the current working directory
p.cwd()
```

```python
# get the home directory
p.home()
```

```python
# join paths
base = Path("data")
file = base / "student.csv"


# NOTE: /  -> means "join paths", not division
```

```python
print(file)
```

```python
# file and folder properties:

# file name
p.name
```

```python
# file extension(suffix)
p.suffix
```

```python
# get the file name without extension
p.stem
```

```python
# get the parent directory
p.parent
```

```python
# get the all parents
p.parents
```
